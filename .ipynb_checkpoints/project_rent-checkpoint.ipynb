# Feature Engineering

```python
# Renaming beds into bhk column
df = df.rename(columns = {'beds':'bhk'})
```

## Creating new features from existing ones

```python
df['area_per_bhk'] = df['area']/df['bhk']
threshold = df['area_per_bhk'].quantile(0.75)
print(threshold)
```

```python
for a in df:
    df['IsLuxury'] = (df[['area_per_bhk']]>threshold).astype(int) 
```

By creating this column **IsLuxury** and **area_per_bhk**,we will get more precise information about the population spread across entire area of the flat and thus infering which ine is ecpensive and will be categorised as luxury apartment

```python
df = df.drop(['house_type'],axis=1)
```

```python
for b in df:
    df['HasView'] = (df[['balconies']]>0).astype(int)
```

By creating column **HasView**, we will get precise information about which flat owns balconies and have a views like seaside view,mountain view etc. 

## Outlier detection and capping

```python
Q1 = df['area'].quantile(0.25)
Q2 = df['area'].quantile(0.75)
iqr = Q2-Q1
min_rate = Q1 - 1.5*iqr
max_rate = Q2 + 1.5*iqr


df['area'] = np.where(
    df['area'] > max_rate,
    max_rate,
    np.where(
        df['area'] < min_rate,
        min_rate,
        df['area']
    )
)
```

```python
Q1r = df['area_rate'].quantile(0.25)
Q2r = df['area_rate'].quantile(0.75)
iqr1 = Q2r-Q1r
min_rate1 = Q1r - 1.5*iqr1
max_rate1 = Q2r + 1.5*iqr1


df['area_rate'] = np.where(
    df['area_rate'] > max_rate1,
    max_rate1,
    np.where(
        df['area_rate'] < min_rate1,
        min_rate1,
        df['area_rate']
    )
)
```

## Train Test split on the dataset

```python
X_train,X_test,Y_train,Y_test = train_test_split(df.drop(['rent'],axis=1),
                                               df['rent'],
                                               test_size=0.2,
                                               random_state=42
                                               )
```

## log transformation of target variables

```python
func_num = FunctionTransformer(func=np.log1p)
```

```python
Y_train_log = func_num.fit_transform(Y_train)
Y_test_log = func_num.transform(Y_test)
```

By transforming the target variable and dropping area_rate column,we avoid data leakage

## Creating a list for categorical columns

```python
cat_cols = X_train.select_dtypes(include='object').columns
cat_cols
```

## Creating a list for numerical columns

```python
num_cols = X_train.select_dtypes(include=['int64','float64']).columns
num_cols
```

## Using Columntransformer and Pipeline
In this section I am going to use **ColumnTransformer** and I am going to apply **log transformation** and **StandardScaler** on **area_rate** and **area**.
After column transformation, I am going to create models , I have chosen regression models:
1. LinearRegression
2. DecisionTreeRegressor
3. RandomForestRegressor
4. GradientBoostingRegressor

Then finally I am going to create four **PipeLines**

```python
preprocessing_num = ColumnTransformer(transformers=
                        [
                        ('ohe',OneHotEncoder(handle_unknown='ignore',drop='first'),cat_cols),
                        ('scale',StandardScaler(),num_cols)
                        ],remainder='drop')
```

```python
model1 = LinearRegression()
model2 = DecisionTreeRegressor(random_state=42,min_samples_split=40,
    min_samples_leaf=30)
model3 = RandomForestRegressor(n_estimators=1000,max_depth=6,min_samples_leaf=50,min_samples_split=40)
model4 = GradientBoostingRegressor(n_estimators=1600,
    learning_rate=0.01,
    max_depth=4,
    subsample=0.6,
    min_samples_leaf=70,
    min_samples_split=60,
    alpha=0.01,
    max_features=0.7,
    random_state=42)
```

We have regularised the models so in order to avoid **overfitting** and have a better **Bias-Variance trade-off**,thus generalising the models

# Model Training

```python
pipe1 = Pipeline(
    [
    ('num',preprocessing_num),
    ('model1',model1)
    ]
)
pipe1.fit(X_train,Y_train_log)
```

```python
pipe2 = Pipeline(
    [
    ('num',preprocessing_num),
    ('model2',model2)
    ]
)
pipe2.fit(X_train,Y_train_log)
```

```python
pipe3 = Pipeline(
    [
    ('num',preprocessing_num),
    ('model3',model3)
    ]
)
pipe3.fit(X_train,Y_train_log)
```

```python
pipe4 = Pipeline(
    [
    ('num',preprocessing_num),
    ('model4',model4)
    ]
)
pipe4.fit(X_train,Y_train_log)
```

# Predicting the output of the test dataset

```python
Y_pred1 = pipe1.predict(X_test)
Y_pred2 = pipe2.predict(X_test)
Y_pred3 = pipe3.predict(X_test)
Y_pred4 = pipe4.predict(X_test)
```

# Model Evaluation Metrics
Before comparing our regression models, let's understand the key metrics used to assess performance:

**R2 Score** (Coefficient of Determination):
1. Measures how well predictions explain variance in actual values (0-1 range)
2. 1.0 = perfect prediction, 0.0 = predicts mean only, <0 = worse than mean
3. Higher is always better

**MAE (Mean Absolute Error)**:
1. Average absolute difference between predictions and actual values
2. Smaller values = better accuracy
3. Intuitive: "average prediction error in original units"

**RMSE (Root Mean Squared Error)**:
1. Penalizes large errors more heavily than MAE (squared errors)
2. Smaller values = better
3. Same units as target (easier interpretation for rent prediction)

**Cross-Validation (CV) Score**:
1. 10-fold CV R² average: Trains on 9/10 data folds, tests on 1/10, repeated 10x
2. Prevents overfitting by testing generalization across data splits
3. Most reliable metric for model selection


```python
# Linear Regression results 
lr_r2 = r2_score(Y_test_log,Y_pred1)
lr_mae = mean_absolute_error(Y_test_log,Y_pred1)
lr_rmse = np.sqrt(mean_squared_error(Y_test_log,Y_pred1))
# Decision Tree Regressor results   
dtr_r2 = r2_score(Y_test_log, Y_pred2)
dtr_mae = mean_absolute_error(Y_test_log, Y_pred2)
dtr_rmse = np.sqrt(mean_squared_error(Y_test_log, Y_pred2))

# Random Forest Regressor results 
rf_r2 = r2_score(Y_test_log,Y_pred3)
rf_mae = mean_absolute_error(Y_test_log,Y_pred3)
rf_rmse = np.sqrt(mean_squared_error(Y_test_log,Y_pred3))

# Gradient Boost Regressor results

gbr_r2 = r2_score(Y_test_log,Y_pred4)
gbr_mae = mean_absolute_error(Y_test_log,Y_pred4)
gbr_rmse = np.sqrt(mean_squared_error(Y_test_log,Y_pred4))



# Kfold
k = 10  
kf = KFold(n_splits=k, shuffle=True, random_state=42)
# Cross Validation scores
cv1 = cross_val_score(pipe1,X_train,Y_train_log,cv=kf,scoring='r2').mean()
cv2 = cross_val_score(pipe2,X_train,Y_train_log,cv=kf,scoring='r2').mean()
cv3 = cross_val_score(pipe3,X_train,Y_train_log,cv=kf,scoring='r2').mean()
cv4 = cross_val_score(pipe4,X_train,Y_train_log,cv=kf,scoring='r2').mean()

# Comparison table
results = pd.DataFrame({
    'Model': ['Linear Regression','Decision Tree','Random Forest','Gradient Boost'],
    'R² Score': [lr_r2,dtr_r2,rf_r2,gbr_r2],
    'MAE': [lr_mae,dtr_mae,rf_mae,gbr_mae],
    'RMSE': [lr_rmse,dtr_rmse,rf_rmse,gbr_rmse],
    'CV' :  [cv1,cv2,cv3,cv4]
})
print(f'The results of the models are \n {results}')
#print(results)
```

# Result of every model:

After analyzing the results given above we conclude that:
1. Gradient Boosting Regressor: It comes out as top performer with results of:
   - Highest R2 score of 0.813572 (81.36 % approximately) which indicates great accuracy
   - Lowest RMSE error of 0.427290 (42.73 %)
   - A good cross validation score (CV score) of 0.805235 (80.52 %)
2. Random Forest Regressor: It has second best score after Gradient Boosting:
   - Second highest R2 score of 0.770783 (77.08 % approximately) 
   - RMSE error of 0.473796 (47.38 % approximately)
   - Highest cross validation score (CV score) of 0.760774 (76.08 % approximately)
3. Linear Regression: It also gives a competitive result:
   - R2 score of 0.757086 (75.71 % approximately) just slightly lower than Random Forest 
   - RMSE error of 0.487747 (48.77 %)
   - A good cross validation score (CV score) of 0.698609 (70% approximately)
4. Decision Tree Regressor : It has a lowest results as compared to other three models:
   - R2 score of 0.731472 (73.15 % approximately).
   - RMSE error of 0.512817 (51.28 %)
   - A good cross validation score (CV score) of 0.710211 (71.02 %)

So, after analyzing the result I came to conclusion that Gradient Boosting gives best result due to high R2 score, Lowest RMSE error and a good CV score due to which it is more generalised, robust model to deploy.
